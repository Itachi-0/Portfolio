```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('salesbigdata.csv', low_memory=False)
```

```python
df
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df,x='age',bins=20,kde=True)
plt.title('Distribution of age')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10,9))
df['payment_method'].value_counts().plot.pie(autopct='%1.2f%%')
plt.title('Payment method distributed')
plt.xlabel('')
plt.show
```
